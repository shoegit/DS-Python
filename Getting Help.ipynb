# Getting Help in Jupyter

For help on Python, you can actually just type help() and run the cell. Try it below. 

In the text box that appears, you can type what you need help about. 

For example, enter `keywords` to get the list of keywords that are used in Python, so that you know that you should not use these as Python variable names. 


```python
help()
```
